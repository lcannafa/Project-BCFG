```python
import pandas as pd
import numpy as np
import scipy as sp 

data = pd.read_excel('C:\\Users\\usuario\\OneDrive\\Documents\\EAFIT\\Semestre9\\Ingenieria del conocimiento\\Datasets Clustering\\Playas.xlsx')
data[['Tipo Suelo','Oleaje','Ocupación','Mar']] = data[['Tipo Suelo','Oleaje','Ocupación','Mar']].stack().rank(method='dense').unstack()
data['Ocupación'] = data['Ocupación'].astype(int)
data[['Tipo Suelo','Oleaje','Mar']]=(data[['Tipo Suelo','Oleaje','Mar']]-data[['Tipo Suelo','Oleaje','Mar']].mean())/data[['Tipo Suelo','Oleaje','Mar']].std()
#data[['Tipo Suelo','Oleaje','Ocupación','Mar']]=(data[['Tipo Suelo','Oleaje','Ocupación','Mar']]-data[['Tipo Suelo','Oleaje','Ocupación','Mar']].min())/(data[['Tipo Suelo','Oleaje','Ocupación','Mar']].max()-data[['Tipo Suelo','Oleaje','Ocupación','Mar']].min())
print("Tipo Suelo",data['Tipo Suelo'].unique())
print("Oleaje",data['Oleaje'].unique())
print("Ocupación",data['Ocupación'].unique())
#17 Alta, 26 Media, 20 Baja
print("Mar",data['Mar'].unique())
```

```python
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

sb.pairplot(data.dropna(), hue='Ocupación',height=4,vars=["Oleaje","Tipo Suelo","Mar"],kind='scatter')
```

```python
X = np.array(data[["Oleaje","Tipo Suelo","Mar"]])
y = np.array(data['Ocupación'])
X.shape
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
#colores=['blue','red','green','blue','cyan','yellow','orange','black','pink','brown','purple']
asignar=[]
for row in y:
    asignar.append('purple')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

Nc = range(1, 10)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
print(score)
plt.plot(Nc,score)
plt.xlabel('Número de  Clusters')
plt.ylabel('Puntaje')
plt.title('Elbow Curve')
plt.show()
```

```python
cost =[] 
for i in range(1, 10): 
    KM = KMeans(n_clusters = i, max_iter = 500) 
    KM.fit(X)
    cost.append(KM.inertia_)
print(cost)    
plt.plot(range(1, 10), cost, color ='g', linewidth ='3') 
plt.xlabel("Valor de K") 
plt.ylabel("Squared Error (Cost)") 
plt.show()
```

```python
kmeans = KMeans(n_clusters=6).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['red','green','blue','orange','cyan','purple']
asignar=[]
for row in labels:
    asignar.append(colores[row])
 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=1000)
```
